```python
x = df["TV"]
y = df["Sales"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
x_train = x_train.to_frame()
ML = LinearRegression()
ML.fit(x_train, y_train)
```

```python
x_train_pred = ML.predict(x_train)
print(x_train_pred)
```

```python
print(y_train)
```

```python
print(ML.coef_)
print(ML.intercept_)
```

```python
x_test = x_test.to_frame()
x_test_pred = ML.predict(x_test)
print(x_test_pred)
```

```python
print(y_test)
```

```python
plt.figure(figsize = (7,5))
plt.scatter(y_train , x_train_pred)
plt.xlabel("orginal values")
plt.ylabel("predicted values")
plt.title("orginal vs predicted values")
plt.show()
```

```python
plt.figure(figsize = (7,5))
plt.scatter(y_test , x_test_pred)
plt.xlabel("orginal values")
plt.ylabel("predicted values")
plt.title("orginal vs predicted values")
plt.show()
```
